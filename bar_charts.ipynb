# Counter
Here we are reading a csv which has the small part of our big dataset and it has the programming languages with which the people who has answered the survey has worked with.
So we will get the most common language out of it and for that we are using the collection module

```python
with open("output.csv", ) as file_:
    csv_reader = csv.DictReader(file_)
    csv_reader = csv.DictReader(file_)
    languages_counter = Counter()

    for row in csv_reader:
        languages_counter.update(row['LanguageHaveWorkedWith'].split(";"))  

languages = []
languages_popularity = []
for items in languages_counter.most_common(15):
    languages.append(items[0])
    languages_popularity.append(items[1])
    
languages.reverse()
languages_popularity.reverse()

plt.barh(languages, languages_popularity)  
plt.title("Most common programming languages")
plt.xlabel("languages_popularity")
plt.tight_layout()

plt.grid()
      
    
```

# Using pandas to read csv

```python
data = pd.read_csv("output.csv")
lang_responses = data['LanguageHaveWorkedWith']
counter = Counter()

for responses in lang_responses:
    
    if pd.notna(responses):
        counter.update(responses.split(";"))
    else:
        pass

language = []
language_popularity = []

for lang in counter.most_common(15):
    language.append(lang[0])
    language_popularity.append(lang[1])
    
language.reverse()
language_popularity.reverse()

plt.barh(language, language_popularity, color='r')  
plt.title("Most common programming languages")
plt.xlabel("languages_popularity")
plt.tight_layout()

plt.grid()

    
```
