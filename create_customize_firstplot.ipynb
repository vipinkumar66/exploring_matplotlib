```python
from matplotlib import pyplot as plt
```

```python
ages = [20, 21, 22, 23, 24, 25, 26] #this is representing the age
dev_y = [20000, 30000, 45600, 57000, 76345,
            89888, 100000] #this is represnting the salary for those developers

plt.plot(ages, dev_y)
plt.show() #this is must if we want to see the graph otherwise it is not possible
```

ADDING LABELS
to format the plots: fmt = [marker][color][line] this is the order and you can refer the documentation part of the matplotlib for the different types 

```python
plt.plot(ages, dev_y, '.b--', label="All developers") #this will present a blue color dotted line
# You can also pass this as color="blue", linestyle= "--"
py_dev = [30000, 40000, 55600, 67000, 86345,
            99888, 110000]
plt.plot(ages, py_dev, marker = "o", color = "r" , label="Python developers") #and this will get the red color line
plt.xlabel("Developer Ages")
plt.ylabel("Salary by USD")
plt.title("Developer salary in USD by ages")
plt.legend(loc="upper left")
```

```python

plt.style.use("ggplot")
plt.xkcd() #this gives a comic type style
plt.plot(ages, dev_y, '.b--', linewidth=3, label="All developers") #this will present a blue color dotted line
# You can also pass this as color="blue", linestyle= "--"
py_dev = [30000, 40000, 55600, 67000, 86345,
            99888, 110000]
plt.plot(ages, py_dev, marker = "o", color = "r" , label="Python developers") #and this will get the red color line
java_dev = [37000, 39000, 55600, 70000, 89345,
            100888, 118000]
plt.plot(ages, java_dev, color= 'k', linewidth=3 , label="Java developer")
plt.xlabel("Developer Ages")
plt.ylabel("Salary by USD")
plt.title("Developer salary in USD by ages")
plt.legend(loc="upper left")
plt.grid(True)
# suppose you are facing any padding issue than at time you can use this command
plt.tight_layout()
```
