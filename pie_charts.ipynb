```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
slices = [40, 60, 90]
labels = ['python', 'java', 'c++']
colors = ['#9ebd9e', '#57dd73', '#c09ede']
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'}, colors=colors)
plt.title("My pie Chart")
plt.tight_layout()
```

```python
import pandas as pd
from collections import Counter
import csv

with open("output.csv", ) as file_:
    csv_reader = csv.DictReader(file_)
    csv_reader = csv.DictReader(file_)
    languages_counter = Counter()

    for row in csv_reader:
        languages_counter.update(row['LanguageHaveWorkedWith'].split(";"))  

languages = []
languages_popularity = []
for items in languages_counter.most_common(15):
    languages.append(items[0])
    languages_popularity.append(items[1])
    
plt.pie(languages_popularity, labels=languages, wedgeprops={"edgecolor":"black"})
```

The above part looks crowdy so we should use bar chart for so much item, so let's include only top 6 languages
We will also look into:
(1) shadow
(2) explode
(3) autopct

```python
# There is a property called explode, it basically focus on the value whcich we want it to show specially.
explode = [0.1, 0, 0, 0, 0, 0]

plt.pie(languages_popularity[:6], labels=languages[:6], wedgeprops={"edgecolor":"black"}, explode=explode,
        shadow=True, autopct="%1.1f%%")
```
